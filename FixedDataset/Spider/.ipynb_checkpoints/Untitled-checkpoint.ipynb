```python
import json
import operator
import os
```

```python
raw_data = json.load(open('train_spider.json'))
raw_data = raw_data + json.load(open('train_others.json'))
tables = json.load(open('tables.json'))
#for line in open('train_spider.json', 'r'):
#    data.append(json.load(line))
```

```python
query_count = {}

for item in raw_data:
    db_id = item['db_id']
    if db_id in query_count:
        query_count[db_id] = query_count[db_id] + 1
    else:
        query_count[db_id] = 1
```

```python
query_count = sorted(query_count.items(),key=operator.itemgetter(1),reverse=True)
```

```python
query_count
```

```python
for item in raw_data:
    db_id = item['db_id']
    if db_id == 'scholar':
        print(item['question'])
```

```python
table_count = {}
attribute_count = {}
for item in tables:
    for i in range(16):
        if query_count[i][0]  == item['db_id']:
            table_count[item['db_id']] = len(item['table_names'])
            attribute_count[item['db_id']] = len(item['column_names'])
```

```python
print(table_count)
print(attribute_count)
```

```python
for i in range(10):
    os.mkdir(str(query_count[i][0]))
    file = open(str(query_count[i][0]) + "/questions.txt", "w+")

    for item in raw_data:
        if(item['db_id'] == query_count[i][0]):
            try:
                file.write(item['question'] + '|||' + item['query'] +'\n')
            except IndexError:
                print('error occured')
                print(item)
                print('end of error')
            
    file.close()
```
